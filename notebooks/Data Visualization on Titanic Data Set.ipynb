# Data Visualization

## Importing Modules

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading Dataset

```python
df = pd.read_csv('Downloads/Titanic-Train-Data.csv')
df
```

# Finding NAN values : [Heat map]

```python
sns.heatmap(df.isnull())
```

# Sex Vs Age using Bar-Graph 

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Age", data=df)
plt.show()
```

# Sex Vs Age Vs Pclass using Bar-Plot 

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Age", hue="Pclass", data=df)
plt.show()
```

# Sex Vs Age Vs Pclass using Violin-Plot 

```python
sns.violinplot(x="Sex", y="Age", hue='Pclass', data=df)
plt.show()
```

# Classifying Age using Distplot

```python
sns.distplot(df['Age'])
plt.show()
```

# Classifying Survived Vs Age Vs Pclass using Bar-Plot 

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Survived", y="Age", hue="Pclass", data=df)
plt.show()
```
